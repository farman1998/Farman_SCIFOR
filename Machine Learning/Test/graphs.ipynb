## **Handling Missing Values.**


```python
#importing pandas library
import pandas as pd
```

```python
#load dataset
df = pd.read_csv('phones.csv')
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
#To drop rows with any missing values
df.dropna(axis=0, inplace=True)
```

```python
#To drop columns with any missing values
df.dropna(axis=1, inplace=True)
```

```python
#Save the cleaned dataset
df.to_csv('phones.csv', index=False)
```

## **Plotting Chart**


```python
#for plotting chart use matplotlib
import matplotlib.pyplot as plt
```

```python
# All Brands
x = df['Model']
y = df['Price']
```

##Bar Chart

```python
# Bar chart
plt.xlabel('Model', fontsize=10)
plt.ylabel('Price($)', fontsize=16)
plt.bar(x, y)
plt.title("Phones Price by Brand")
plt.xticks(rotation = 90)
plt.show()
```

##Pie Chart

```python
plt.pie(y, labels=x, radius=1.2,autopct='%0.01f%%', shadow=True, explode=[.05,.2,.05,.2,.05,.2,.05])
```

##Line Chart

```python
plt.xlabel('Model', fontsize=18)
plt.ylabel('Price($)', fontsize=16)
plt.scatter(x, y)
plt.plot(x, y)

plt.show()
```

#Histogram

```python
# Plot a histogram of a single column in the DataFrame
df.hist(column='Price')

# Set the title and axis labels
plt.title('Price of phone over the brand')
plt.xlabel('Price of phones')
plt.ylabel('Frequency')

# Display the histogram
plt.show()
```

#**Pandas**
####Primary Use: Data manipulation and analysis.

**Key Features:**

* Provides data structures like DataFrame and Series, which are powerful tools for handling structured data.
* Offers functionalities for reading and writing data between in-memory data structures and various file formats (e.g., CSV, Excel, SQL databases).
* Supports data cleaning, reshaping, merging, and slicing operations.
* Facilitates data aggregation and time series functionalities.

**Example Use Case:** Loading data from various sources (files, databases), performing data cleaning and preprocessing, and conducting exploratory data analysis.

#**NumPy**
####Primary Use: Numerical computing and array operations.

**Key Features:**

*Provides support for large, multi-dimensional arrays and matrices.
*Offers a wide range of mathematical functions for array manipulation.
*Efficiently handles numerical computations, making it essential for scientific computing tasks.
*Integrates with other libraries for linear algebra, Fourier transform, and random number generation.

**Example Use Case:** Computing statistical metrics (mean, median, standard deviation) of numerical data, matrix operations, and mathematical transformations.

#**Seaborn**
####Primary Use: Statistical data visualization.

**Key Features:**
* Built on top of Matplotlib, providing a high-level interface for drawing attractive and informative statistical graphics.
* Offers functions for visualizing distributions, regressions, categorical variables, and more complex plots.
* Supports customization of plot aesthetics and themes.
* Simplifies the creation of complex multi-plot grids.

**Example Use Case:** Creating visually appealing statistical plots such as histograms, scatter plots, box plots, and heatmaps for data exploration and presentation.

#**Matplotlib**
####Primary Use: Basic plotting library for creating 2D graphics.

**Key Features:**
* Provides a low-level interface for creating static, animated, and interactive visualizations.
* Supports various plot types, including line plots, bar charts, histograms, scatter plots, and more.
* Allows fine-grained control over every aspect of a plot's appearance.
Integrates well with other libraries like NumPy and Pandas for data visualization.

**Example Use Case:** Generating publication-quality plots for data analysis, scientific research, and presentations.

#**OS Library (os module)**
####Primary Use: Operating system (OS) interface.

**Key Features:**
* Provides functions for interacting with the operating system, such as manipulating files and directories.
* Enables executing shell commands, accessing environment variables, and managing processes.
* Supports platform-independent path manipulation and file system operations.

**Example Use Case:** Automating file operations, managing directories, executing system commands, and handling paths in a cross-platform manner.
