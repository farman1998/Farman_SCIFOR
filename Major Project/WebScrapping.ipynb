#Web Scraping from Real Website

```python
#import libraries
import requests
from bs4 import BeautifulSoup
```

```python
#set webpage in which we scrap data
web = requests.get("https://www.geeksforgeeks.org/what-is-data-science/")
```

```python
#cheack the status of web
web.status_code
```

```python
#display web content
web.content
```

```python
#convert content in HTML form
soup = BeautifulSoup(web.content,"html.parser")
```

```python
# print data in HTML form
print(soup.prettify())
```
